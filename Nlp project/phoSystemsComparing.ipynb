```python
import jellyfish
import pandas as pd
```

```python
similarites_results = {}

similarites_results['system_name'] = []
similarites_results['similarity'] = []
```

```python
def similarity(dic):
    summ = 0
    count = 0
    for k,v in dic.items():
        summ += jellyfish.jaro_distance(k, v)
        count += 1
    return summ/count
```
